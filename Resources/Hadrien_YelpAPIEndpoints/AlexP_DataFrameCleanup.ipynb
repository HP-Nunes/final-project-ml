```python
# Drop unwanted columns
select_columns = df_csv.drop(['alias','display_phone', 'distance', 'id', 'image_url', 'phone', 'transactions', 'url'], axis=1)
select_columns.head()
```

```python
# Clean up column titles
renamed_df_csv = select_columns.rename(columns={"alias":"Alias", "categories":"Categories", 'coordinates': 'Coordinates', 'is_closed': 'Is Closed', 'location': 'Location', 'name': 'Name', 'price': 'Price', 'rating': 'Rating', 'review_count': 'Review Count'})
renamed_df_csv.head()
```

```python
# Change dollar symbols to numbers
price_fixed_df = renamed_df_csv.replace(['$', '$$', '$$$', '$$$$'], [1, 2, 3, 4]).head()
price_fixed_df
```

```python
price_fixed_df['lat'], price_fixed_df['long'] = price_fixed_df['Coordinates'].str.split(', ', 1).str
price_fixed_df['lat'] = price_fixed_df['lat'].str.strip("{'latitude':")
price_fixed_df['long'] = price_fixed_df['long'].str.strip("{'longitude':}")
price_fixed_df
```

```python
price_fixed_df['Location'][0]
```

```python
# eval(price_fixed_df['Location'][0])['display_address'][0]
```

```python
price_fixed_df['Loc'] = eval(price_fixed_df['Location'][0])['display_address'][0]
price_fixed_df
```

```python
# Turn Location string into dictionary
s = price_fixed_df.loc[0, 'Location']
```

```python
s
```

```python
# Create new variable for the dictionary; split the string based on the square bracket; choose the second piece using the index of 1, rather than 0; take everything other than the last three characters in the string.
s_new = s.split("[")[1][:-3]
s_new
```

```python
# Replace/remove the parenthesis
s_new.replace("'", "")
```

```python
# Function to pull the information and apply to new column, upon each row of information
def clean_up_location(row):
    s_new = row.split("[")[1][:-3]
    return s_new.replace("'", "")

price_fixed_df.loc[:, 'Address'] = price_fixed_df.Location.apply(clean_up_location)
```
