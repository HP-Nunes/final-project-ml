```python
#dependencies
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
#Set path
registered_business_path = os.path.join('..','Raw', 'Map_of_Registered_Business_Locations.csv')

#Read in csv file as dataframe
registered_business_df = pd.read_csv(registered_business_path)

#List of columns
registered_business_df.columns.values
```

```python
#number of rows in dataset
len(registered_business_df)
```

```python
registered_business_df.head(5)
```

```python
#Pull the columns that we need
df = registered_business_df[["Location Id", "Ownership Name", "DBA Name",  "NAICS Code Description", "LIC Code Description", "Street Address", "State", "City", 
                                   "Source Zipcode", "Business Start Date", "Business End Date", 
                                   "Neighborhoods - Analysis Boundaries", "Business Location", "Supervisor District"]]
df.head(5)
```

```python
#rename the columns and remove whitespaces
final_df = df.rename(columns={'Location Id': 'location_id', 'Ownership Name': 'ownership_name', 'DBA Name':'business_name', 
                             'NAICS Code Description':'category', 'LIC Code Description':'lic_code', 'Street Address':'street_address', 'Source Zipcode':'zipcode',
                              'Business Start Date':'business_start_date', 'Business End Date': 'business_end_date',
                             'Neighborhoods - Analysis Boundaries': 'neighborhoods', 'Business Location':'business_location', 'Supervisor District':'supervisor_district'})
final_df.head(5)
```

```python
#get unique list of category
final_df.category.unique()
```

```python
#number of rows after dropping values with null values
len(final_df)
```

```python
#convert date columns from str -> date
# pd.to_datetime(final_df['business_end_date'])
# pd.to_datetime(final_df['business_start_date']).dtype

pd.to_datetime(final_df['business_end_date'], format="%m/%d/%Y")
pd.to_datetime(final_df['business_start_date'], format="%m/%d/%Y")

#Add a new column to calculate lifespan
final_df['Lifespan'] = pd.to_datetime(final_df['business_end_date'], format="%m/%d/%Y") - pd.to_datetime(final_df['business_start_date'], format="%m/%d/%Y")

#Extract the integer only (e.g. 508 days -> 508.0)
final_df['Lifespan'] = final_df['Lifespan'].dt.days

#reset the index
final_df = final_df.reset_index(drop=True)

final_df.head(20)
```

```python
#Add a column to indicate whether the business closed or not
final_df['closed'] = np.where(final_df.Lifespan.isnull(),"no","yes")

#Add a column to calculate the Age of a business if it is still open
final_df['age_if_opened'] = pd.to_datetime('today') - pd.to_datetime(final_df['business_start_date'], format="%m/%d/%Y")

#Extract the integer only (e.g. 508 days -> 508.0)
final_df['age_if_opened'] = np.where(final_df.closed == 'no', final_df['age_if_opened'].dt.days, np.nan)

final_df.head(5)
```

```python
#Export cleaned data to csv
path = os.path.join('Cleaned Datasets', 'cleaned_registered_businesses.csv')
final_df.to_csv(path, index=False)
```
