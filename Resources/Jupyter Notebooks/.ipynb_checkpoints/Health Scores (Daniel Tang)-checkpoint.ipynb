```python
#merge the files with left join
final_df = businesses_df.merge(inspections_df, left_on='business_id', right_on='business_id', how='left')
#final_df = final_df.merge(violations_df, left_on=['business_id','date'], right_on=['business_id','date'], how='left')
final_df = final_df.sort_values(by=['name','date'], ascending=True)
final_df.head(100)
```

## Rolling Mean 

```python
# final_df['MA'] = final_df.rolling(window=3)['score'].mean()
final_df.columns.values
```

```python
# df['x'] = df.groupby('id')['x'].rolling(2).mean().reset_index(0,drop=True)

final_df['MA'] = final_df.groupby(['name'])['score'].rolling(2, min_periods=1).mean().reset_index(0, drop=True)
final_df.head(100)
```

```python
#format the name and address so we can group by it
final_df['formatted_name'] = final_df['name'].str.split().str[0] + " " + final_df['name'].str.split().str[1]
final_df['formatted_address'] = final_df['address'].str.split().str[0] + " " + final_df['address'].str.split().str[1]

#drop the null formatted addresses and names
final_df = final_df.dropna(subset=['formatted_name'])
final_df = final_df.dropna(subset=['formatted_address'])


final_df = final_df[final_df['year'] == final_df.groupby(['formatted_name','formatted_address'])['year'].transform(max)]

final_df.head(100)
```

```python
#Export cleaned data to csv
path = os.path.join('Cleaned Datasets', 'aggregated_health_scores_ma.csv')
final_df.to_csv(path, index=False)
```
