```python
# Dependencies
import pandas as pd
import os
```

```python
# Store filepath in a variable
file_one = os.path.join("../Raw", "Buyout_agreements.csv")
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")
```

```python
# Show 5 rows and the header 
file_one_df.head()
```

```python
# Show mulitple specific columns--note the extra brackets
cleaned_buyouts = file_one_df[["Analysis Neighborhood","Block Address","ZIP Code","the_geom","Buyout Agreement Date","Buyout Amount", "Number of Tenants"]]
cleaned_buyouts.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
cleaned_buyouts.to_csv("Cleaned Datasets/Buyout_Agreements_cleaned.csv", index=False, header=True)
```
